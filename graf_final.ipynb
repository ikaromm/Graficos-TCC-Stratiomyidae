```python
import pandas as pd 
import matplotlib as plt
import matplotlib.pyplot as plt
import numpy as np

plt.figure()
```

```python
dados = pd.read_excel("excel/dados.xlsx")
abund = pd.read_excel("excel/tab2.xlsx", sheet_name="abund")
riq = pd.read_excel("excel/tab2.xlsx", sheet_name="riq")

# Create a single figure and ax object
fig, ax = plt.subplots()


# Set the width of each bar
bar_width = 0.35  # Adjust this value to control the width of the bars

# Create the first bar plot
bar1_positions = [0, 1.05, 2, 3, 4, 5, 6]
riq_values = riq.iloc[0, 1:]

for i, (position, value) in enumerate(zip(bar1_positions, riq_values)):
    bar1 = ax.bar(position, value, bar_width, color = 'lightgreen')

    # Add the value on top of the bar
    ax.text(position, value + 0.1, str(value), ha='center', va='bottom')

# Create the second bar plot
bar2_positions = [-0.35, 0.7, 1.65, 2.65, 3.65, 4.65, 5.65]
abund_values = abund.iloc[0, 1:]

for i, (position, value) in enumerate(zip(bar2_positions, abund_values)):
    bar2 = ax.bar(position, value, bar_width, color='lightslategray')

    # Add the value on top of the bar
    ax.text(position, value + 0.1, str(value), ha='center', va='bottom')



# Set labels and title

ax.set_ylabel('temperatura (C), preciptao (mm) e umidaderelativa(%)')
ax.set_title(' ')

# Calculate the center of the bars
center_positions = [(bar1_position + bar2_position) / 2 for bar1_position, bar2_position in zip(bar1_positions, bar2_positions)]

# Set the x-axis tick positions to the center of each bar
ax.set_xticks(center_positions)

# Set the x-axis tick labels
ax.set_xticklabels(riq.columns[1:])

#
plt.yticks([0, 50,100,100,150,200,250,300,350])

###TEMP

# Read the temperature data and month labels
temperature_data = dados.iloc[0, 1:]
prec_data = dados.iloc[1, 1:]
humidade_data = dados.iloc[2, 1:]
month_labels = dados.columns[1:]  # Assuming the first column is the index

# Convert the temperature data to numeric data type
temperature_data = pd.to_numeric(temperature_data)

ax.plot(temperature_data, label='Temperatura',linewidth=3.0, color = "darkorange")
ax.plot(prec_data, label='Preciptao',linewidth=3.0, color = "red")
ax.plot(humidade_data, label='Humidade',linewidth=3.0, color = "mediumslateblue")
# Add the legend
#ax.legend(loc='upper right', frameon=False)
bar1.set_label('Riqueza')
bar2.set_label('Abundancia')
ax.set_xticklabels(riq.columns[1:], rotation=45)
ax.legend()

#plt.grid(axis = "y")

# Show the plot
plt.show()
```
