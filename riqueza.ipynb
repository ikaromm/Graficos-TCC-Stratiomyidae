```python
import pandas as pd 
import matplotlib as plt
import matplotlib.pyplot as plt
import numpy as np

plt.figure()
```

```python
rique = pd.read_excel("excel/rique.xlsx")

j= 0
familia = []
count = []	
for i in rique["subfamilia"]:

    x = i
    y = rique["qnt"][j]
    j += 1
    familia.append(x)
    count.append(y)
    
plt.figure(figsize=(12, 3))

b = 'blue'
g = 'deeppink'
r = 'red'
c =  'cyan'
m = 'magenta'
y = 'green'
k = 'brown'
w = 'orange'

lista = [b, g, r, c, m, y, k, w]

for i in range(len(familia)):
    plt.bar(familia[i], count[i], width=.2, color = lista[i])

plt.grid(False)
plt.title("Riqueza das espécies das subfamílias de Stratiomyidae")
plt.ylabel("Número de espécies")

plt.show()

```
